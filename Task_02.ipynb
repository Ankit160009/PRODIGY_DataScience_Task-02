```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

file_path = '/content/auto-mpg.csv'
auto_mpg_data = pd.read_csv(file_path)

# Convert 'horsepower' to numeric, handling missing values
auto_mpg_data['horsepower'] = pd.to_numeric(auto_mpg_data['horsepower'], errors='coerce')
auto_mpg_data['horsepower'].fillna(auto_mpg_data['horsepower'].median(), inplace=True)

# 2. Exploratory Data Analysis (EDA)
# Distribution of MPG
plt.figure(figsize=(8, 6))
sns.histplot(auto_mpg_data['mpg'], kde=True, color='skyblue')
plt.title('Distribution of MPG')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()

# MPG over model years
plt.figure(figsize=(8, 6))
sns.lineplot(x='model year', y='mpg', data=auto_mpg_data, marker='o', color='green')
plt.title('MPG Trends Over Model Years')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.xticks(np.arange(70, 83, 1))  # Model years from 1970 to 1982
plt.show()

# 3. Explore Relationships Between Variables
# MPG vs. Weight
plt.figure(figsize=(8, 6))
sns.scatterplot(x='weight', y='mpg', data=auto_mpg_data, color='red')
plt.title('MPG vs. Weight')
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.show()

# MPG vs. Horsepower
plt.figure(figsize=(8, 6))
sns.scatterplot(x='horsepower', y='mpg', data=auto_mpg_data, color='blue')
plt.title('MPG vs. Horsepower')
plt.xlabel('Horsepower')
plt.ylabel('MPG')
plt.show()

# Correlation Heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = auto_mpg_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Between Variables')
plt.show()
```

**Interpretation and Insights:-**

Data Cleaning involved converting the horsepower column to numeric and filling missing values with the median, making the dataset ready for analysis.

Exploratory Data Analysis (EDA) revealed the distribution of MPG across vehicles, showing a right skew in MPG values. The trend of MPG over model years illustrated improvements in fuel efficiency over time, likely reflecting advancements in automotive technology and changing regulations.

**Relationship Exploration:**

1. The scatterplot of MPG vs. weight indicated a negative relationship, suggesting that heavier vehicles tend to have lower MPG.
2. The scatterplot of MPG vs. horsepower also showed a negative relationship, implying that cars with higher horsepower, which often equates to higher engine power, tend to be less fuel-efficient.
3. The correlation heatmap provided a quick visual summary of how various attributes relate to each other and to MPG, confirming the negative relationships with weight and horsepower and potentially uncovering other areas for deeper investigation.
